## Type of Plots

Line Plots

```python
x = [1,2,3,4,5]
y = [10,20,25,30,40]
```

Bar Chart

### Subplots & Legends

Legend

```python
days = [1,2,3,4,5]
original_price = [10,20,25,30,40]
predicted_price = [8,21,25,28,38]
```

```python
plt.plot(days, original_price);
plt.plot(days, predicted_price);
plt.legend(['Original Price', 'Predicted Price']);
plt.title('Plot with Legends');
plt.xlabel('Weekdays');
plt.ylabel('Price');
```

```python
# Another way
plt.plot(days, original_price, 'y-o', label='Original Price'); # y = yellow (-o is datapoint)
plt.plot(days, predicted_price, 'g-o', label='Predicted Price'); # g = green (-o is datapoint)
plt.legend()
plt.title('Plot with Legends');
plt.xlabel('Weekdays');
plt.ylabel('Price');
```

Subplots

```python
# First subplot - Original Price
plt.subplot(3,1,1); # no.of rows, columns, index grid
plt.plot(days, original_price, 'b-o', label='Original Price');
plt.title('Original Prices Over Time');
plt.xlabel('Weekdays');
plt.ylabel('Price');
plt.legend();

# Second subplot - Predicted Price
plt.subplot(3,1,2);
plt.plot(days, predicted_price, 'r-o', label='Predicted Price');
plt.title('Predicted Prices Over Time');
plt.xlabel('Weekdays');
plt.ylabel('Price');
plt.legend();

# Third Subplot - Difference between Original and Predicted Price
price_difference = [o - p for o, p in zip(original_price, predicted_price)]
plt.subplot(3,1,3);
plt.bar(days, price_difference, color='purple', label='Price Diffrence');       
plt.title('Difference between Original and Predicted Price');
plt.xlabel('Days');
plt.ylabel('Price Difference');
plt.legend();

plt.tight_layout() # it will adjust the size or padding of the chart
```

### Saving the figure

```python
plt.plot(days, original_price, 'b-o', label='Original Price'); 
plt.plot(days, predicted_price, 'r-o', label='Predicted Price');
plt.legend()
plt.title('Plot with Legends');
plt.xlabel('Weekdays');
plt.ylabel('Price');

plt.savefig('save_plot.png', dpi=300) # save the image
```
